```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("C:\\Users\\guilh_000\\Documents\\Titanic-Competition\\train.csv")
test = pd.read_csv("C:\\Users\\guilh_000\\Documents\\Titanic-Competition\\test.csv")
```

```python
train.head(20)
```

```python
train.drop(['Name', 'Cabin'], axis=1, inplace=True)
train.head(20)
```

```python
train.drop(['Ticket'], axis=1, inplace=True)
train.head()
```

```python
train['Age'].mean()

#fill NaN values with mean value
```

```python
train.info()
```

```python
train['Age'] = train['Age'].fillna(train['Age'].mean())
train.head(20)
```

```python
train['Embarked'] = train['Embarked'].fillna('S')
train.info()

#fill NanN values with most frequent value 'S'
```

```python
dummies = []
cols = ['Pclass', 'Sex', 'Embarked']
for col in cols:
    dummies.append(pd.get_dummies(train[col]))
```

```python
titanic_dummies = pd.concat(dummies, axis=1)
```

```python
titanic_dummies.head()
```

```python
train = pd.concat((train,titanic_dummies), axis=1)
train.head()
```

```python
train = train.drop(['Pclass', 'Sex', 'Embarked'], axis=1)
train.head()
```

```python
train.columns
```

```python
X = train[[ 'Age',       'SibSp',
             'Parch',        'Fare',             1,             2,
                   3,      'female',        'male',           'C',
                 'Q',           'S']]
```

```python
y = train.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn import svm
clf = svm.SVC(C= 1, kernel='linear')
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_recall_fscore_support, classification_report

y_pred = clf.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression(random_state=0)
lrc.fit(X_train, y_train)
```

```python
y_pred = lrc.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(max_depth=5, random_state=0)
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
