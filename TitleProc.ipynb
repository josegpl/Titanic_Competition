```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("C:\\Users\\guilh_000\\Documents\\Titanic-Competition\\train.csv")
test = pd.read_csv("C:\\Users\\guilh_000\\Documents\\Titanic-Competition\\test.csv")
```

```python
train.head(20)
```

```python
train.info()
```

```python
train.drop(['Ticket', 'Cabin'], axis=1, inplace=True)
train.info()
```

```python
train['Age'] = train['Age'].fillna(train['Age'].mean())
train['Embarked'] = train['Embarked'].fillna('S')
train.info()
```

```python
train['FamilySize'] = train['SibSp'] + train['Parch'] + 1
train.head()
```

```python
# Using expression pattern to extract the Title of the passenger
train['Title'] = train.Name.str.extract(' ([A-Za-z]+)\.', expand=False)

# Changing to common category
train['Title'] = train['Title'].replace(['Dr', 'Rev', 'Col', 'Major', 'Countess', 'Sir', 'Jonkheer', 'Lady', 'Capt', 'Don'], 'Others')
train['Title'] = train['Title'].replace('Ms', 'Miss')
train['Title'] = train['Title'].replace('Mme', 'Mrs')
train['Title'] = train['Title'].replace('Mlle', 'Miss')
train.head(20)
```

```python
dummies = []
cols = ['Pclass', 'Sex', 'Embarked', 'Title']
for col in cols:
    dummies.append(pd.get_dummies(train[col]))
```

```python
titanic_dummies = pd.concat(dummies, axis=1)
titanic_dummies.head()
```

```python
train = pd.concat((train,titanic_dummies), axis=1)
train.head()
```

```python
train = train.drop(['Pclass', 'Name', 'Sex', 'Embarked'], axis=1)
train.head()
```

```python
train = train.drop(['PassengerId'], axis=1)
train.head()
```

```python
train.columns
```

```python
X = train[[       'Age',      'SibSp',      'Parch',       'Fare',
       'FamilySize',            1,            2,            3,
           'female',       'male',          'C',          'Q',          'S',
           'Master',       'Miss',         'Mr',        'Mrs',     'Others']]
```

```python
X.head()
```

```python
y = train.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.svm import LinearSVC
clf = LinearSVC(C=100.0,random_state=0)
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_recall_fscore_support, classification_report

y_pred = clf.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(max_depth=5, random_state=0)
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression(random_state=0)
lrc.fit(X_train, y_train)
```

```python
y_pred = lrc.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
