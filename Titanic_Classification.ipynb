```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("C:\\Users\\guilh_000\\Documents\\Titanic-Competition\\train.csv")
test = pd.read_csv("C:\\Users\\guilh_000\\Documents\\Titanic-Competition\\test.csv")
```

```python
train['Cabin'].replace(np.nan, ' ', inplace=True)
train.head(20)
```

```python
train.dropna(subset=['Age'], inplace=True)
train.dropna(subset=['Embarked'], inplace=True)
train.dropna(subset=['SibSp'], inplace=True)
train.dropna(subset=['Parch'], inplace=True)
train.dropna(subset=['Fare'], inplace=True)
train.dropna(subset=['Sex'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le_X=LabelEncoder()
train['Sex'] = le_X.fit_transform(train['Sex'])
train.head(20)

#male=1
#female=0
```

```python
le_X=LabelEncoder()
train['Embarked'] = le_X.fit_transform(train['Embarked'])
train.head(20)

#S =2
#C =0
#Q =1
```

```python
import seaborn as sns; sns.set(font_scale=1.2)
sns.lmplot('Parch', 'Age', data=train, hue='Survived',
           palette='Set1', fit_reg=False, scatter_kws={"s": 70});
```

```python
train.columns
```

```python
X = train[['Pclass', 'Sex', 'Age', 'SibSp',
       'Parch', 'Fare']]

#not using 'survived', 'name', 'ticket' 'cabin' 'embarked' columns
```

```python
X.head(20)
```

```python
y = train.iloc[:,1]
```

```python
y.head(20)
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.25,random_state=0)
```

```python
from sklearn import svm
clf = svm.SVC(C= 0.1, kernel='linear')
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_recall_fscore_support, classification_report

y_pred = clf.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression(random_state=0)
lrc.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_recall_fscore_support, classification_report
y_pred = lrc.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(6,6,6), activation='relu', solver='adam', max_iter=500)
mlp.fit(X_train,y_train)
```

```python
y_pred = mlp.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
